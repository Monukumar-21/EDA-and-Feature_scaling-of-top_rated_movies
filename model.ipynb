```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv('Top_Rated_Movies.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
df.head()
```

```python
df.tail()
```

```python
df.isnull().sum()
```

```python
df = df.dropna(subset=['release_date'])
```

```python
df.isnull().sum()
```

```python
df['year'] = df['release_date'].str.split('-').str[0]
df['month'] = df['release_date'].str.split('-').str[1]
df['day'] = df['release_date'].str.split('-').str[2]
```

```python
df.head()
```

```python
df.drop('release_date',axis=1,inplace=True)
```

```python
df.head()
```

```python
df.info()
```

```python
df.columns
```

```python

df['year'] = df['year'].astype(int)
df['month'] = df['month'].astype(int)
df['day'] = df['day'].astype(int)
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
df_scaled=df.copy()
features = ['vote_average', 'vote_count', 'popularity']
df_scaled[features]=scaler.fit(df[features])
```

```python
sns.pairplot(df)
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['vote_average'], bins=30, kde=True, color='teal')
plt.title('Distribution of Movie Ratings')
plt.xlabel('Vote Average')
plt.ylabel('Frequency')
plt.savefig('rating_distribution.png')
```

```python
plt.figure(figsize=(12, 6))
top_10 = df.nlargest(10, 'popularity')
sns.barplot(x='popularity', y='title', data=top_10, palette='magma')
plt.title('Top 10 Most Popular Movies')
plt.tight_layout()
plt.savefig('top_popularity.png')
```

```python
plt.figure(figsize=(12, 6))
yearly_trend = df['year'].value_counts().sort_index()
plt.plot(yearly_trend.index, yearly_trend.values, color='red', marker='o')
plt.title('Number of Movie Releases per Year')
plt.xlabel('Year')
plt.ylabel('Count')
plt.savefig('release_trend.png')
```

```python
plt.figure(figsize=(8, 6))
numeric_cols = df[['vote_average', 'vote_count', 'popularity', 'year']]
sns.heatmap(numeric_cols.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Feature Correlation Matrix')
plt.savefig('correlation_heatmap.png')
```

```python
df_scaled.to_csv('cleaned_and_scaled_movies.csv', index=False)

print("Data cleaning, EDA, and Scaling complete!")
print(f"Final dataset shape: {df_scaled.shape}")
print("Files saved: cleaned_and_scaled_movies.csv and several PNG plots.")
```
